# Utilizar el modelo yolosegm descargado para la segmetacion

```python
from ultralytics import YOLO

# Cargar modelo YOLOv8 segmentación
model = YOLO('yolov8s-seg.pt')
```

## Verificar que se este ejecutando con gpu el codigo

```python
import torch

# Verificar si hay una GPU disponible
print("Dispositivo actual:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU")

# Configurar dispositivo (GPU)
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device)
```

## Primer entrenamiento realizado

```python
results = model.train(
    data="/kaggle/working/datasetfinal/data.yaml",
    task='segment',
    epochs=30,              # Aumentado
    imgsz=640,
    batch=4,                # Perfecto para Kaggle
    workers=2,
    patience=8,             # Ajustado para parada temprana
    save=True,
    project="runs/train",
    name="exp_seg",
    save_period=3           # Guarda cada 5 épocas para ahorrar espacio
)
```

# Utilizar el modelo con finetunin (Ajuste fino)

```python
from ultralytics import YOLO

# Cargar modelo YOLOv8 segmentación
model = YOLO('/kaggle/working/best6.pt')
```

# usar la gpu para hacer mas rapido el entrenamiento

```python
import torch

# Verificar si hay una GPU disponible
print("Dispositivo actual:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU")

# Configurar dispositivo (GPU)
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device)
```

# Entrenar el modelo preentrenado con ajuste fino

```python
results = model.train(
    data="/kaggle/working/datasetfinal/data.yaml",
    task='segment',
    epochs=50,              # Más épocas para convergencia
    imgsz=640,
    batch=4,
    workers=2,
    patience=15,            # Paciencia mayor para EarlyStopping
    save=True,
    project="runs/train",
    name="exp_seg_improved",
    save_period=3,          # Guarda menos seguido para ahorrar espacio

    # === AJUSTES PARA MEJORAR LA SEGMENTACIÓN ===
    lr0=0.001,              # Reducir tasa de aprendizaje inicial para fine-tuning
    weight_decay=0.0005,    # Regularización L2 para evitar overfitting
    box=0.05,
    cls=0.5,
    dfl=1.0
)
```
